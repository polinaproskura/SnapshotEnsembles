```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 1
st = 1
n = 0
x_arr, y_arr = np.linspace(0, 3.1, 30), []
for x in x_arr:
    print(x, end=' ')
    if x >= n + 1:
        n = n + 1
    print(n, end=' ')
    if x < n:
        continue
    if x >= n and x < n + 0.5:
        print("min", end=' ')
        y = 2 * (b - a) - 2 * n * b
        y_arr.append(y)
    elif x < n + 1 and x >= n + 0.5:
        print("max", end=' ')
        y = 2 * (a - b) + (2 * n  + 2) * b
        y_arr.append(y)
    print()
```
